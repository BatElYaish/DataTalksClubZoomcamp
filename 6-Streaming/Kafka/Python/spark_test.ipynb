```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("LocalNotebookToDockerSpark") \
    .config("spark.master", "spark://localhost:7077") \
    .config("spark.driver.host", "host.docker.internal") \
    .config("spark.driver.bindAddress", "0.0.0.0") \
    .config("spark.submit.deployMode", "client") \
    .config("spark.driver.memory", "2g") \
    .config("spark.executor.memory", "2g") \
    .config("spark.driver.port", "40919") \
    .config("spark.blockManager.port", "40920") \
    .config("spark.port.maxRetries", "32") \
    .config("spark.dynamicAllocation.enabled", "false") \
    .getOrCreate()

# Test the connection
df = spark.createDataFrame([(1, "test"), (2, "test2")], ["id", "value"])
df.show()
```
